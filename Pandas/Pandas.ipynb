```python
import numpy as np 
import pandas as pd 
```

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df,values='Sales',index = 'Region', columns="Product")
```

```python
pd.pivot_table(df,values = "Sales",index = 'Region',columns="Product",aggfunc = 'median')
```

```python
pivot2 = pd.pivot_table(df, values=['Sales', 'Units'], index='Region', columns='Product')
pivot2
```

```python
def extract_episodes(txt):
    check = False
    data = ""
    for i in txt:
        if i == ')':
            check=False
            return data
        if check == True:
            data += i
        if i == '(':
            check=True
            
        
            
        
```

```python
def extraction_time(txt):
    check=False
    data = ""
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1, i+20):
                data += txt[j]
            return data     
            
```

```python
df3['Title'].apply(extraction_time)
```

```python
df3['Total Time'] = df3['Title'].apply(extraction_time)
```

```python
df3.head()
```

```python
def calculate_months(duration):
    start, end = duration.split(' - ')
    
    start_date = pd.to_datetime(start, format='%b %Y')
    end_date = pd.to_datetime(end, format='%b %Y')
    
    months = ((end_date.year - start_date.year) * 12 +
              (end_date.month - start_date.month) + 1)
    
    return months

df3['Months'] = df3['Total Time'].apply(calculate_months)
```
