```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Matplotlib Basics


```python
x = np.linspace(0,5,11)
x
y=x**2
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

# Subplots in matplotlib


```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

# object oriented Ways

```python
fig = plt.figure()
axis1=fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2=fig.add_axes([0.1,0.1,0.8,0.8])
axis2.plot(x,y)
```
